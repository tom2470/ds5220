```python
#Write a simulation study to show the effect of number of features to the linear regression model MSE.
#You can set the number of observation as 100, and the range of the features is from 5 to 100.
#Make a plot to show the change of MSE with the number of features.
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd
linmse=[]
mylist=np.arange(5,100,5)
for i in mylist:
  df = pd.DataFrame()
  for k in range(i):
    x= np.random.normal(5, 1, 100)
    df[k]=x
  X=df
  y = np.random.normal(5, 1, 100)
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4400)
  model = LinearRegression()
  model.fit(X_train, y_train)
  # Predict the target variable for the test set (optional)
  y_pred = model.predict(X_test)
  # Calculate the Mean Squared Error (MSE) for the test set (optional)
  mse = mean_squared_error(y_test, y_pred)
  linmse.append(mse)


```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 4))
plt.plot(mylist, linmse, label='MSE', alpha=0.5)
plt.xlabel('features')
plt.ylabel('Accuracy')
plt.title('Accuracy vs features for LR')
plt.legend()
```
